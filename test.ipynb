```python
import requests
import threading

url = "http://10.199.34.65:2569"
num_requests = 10  # จำนวน Thread ที่จะรันพร้อมกัน

def send_request():
    while True: # ยิงวนไปเรื่อยๆ
        try:
            response = requests.get(url)
            print(f"Status: {response.status_code}")
        except Exception as e:
            print(f"Error: {e}")

threads = []
for i in range(num_requests):
    t = threading.Thread(target=send_request)
    t.daemon = True # ให้จบการทำงานเมื่อปิดโปรแกรมหลัก
    threads.append(t)
    t.start()

# กด Ctrl+C เพื่อหยุด
for t in threads:
    t.join()
```
